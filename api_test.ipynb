```python
from ossapi import Ossapi, BeatmapsetSearchMode, Cursor
import os
from dotenv import load_dotenv

load_dotenv()
api = Ossapi(os.getenv("CLIENT_ID"), os.getenv("CLIENT_SECRET"))
```

```python
num_beatmaps = 300
num_saved = 0
cursor = None
beatmaps_id = []

while num_saved < num_beatmaps:
    beatmaps = api.search_beatmapsets(mode=BeatmapsetSearchMode.TAIKO, cursor=cursor)
    cursor = beatmaps.cursor
    
    cur_beatmaps_id = [beatmapset.id for beatmapset in beatmaps.beatmapsets]
    beatmaps_id += cur_beatmaps_id
    
    print(len(beatmaps.beatmapsets))
    num_saved += len(beatmaps.beatmapsets)

print(beatmaps_id)
```
