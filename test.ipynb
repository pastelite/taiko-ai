```python
import torchaudio
import matplotlib
import matplotlib.pyplot as plt
import os
import torch
```

```python

path = os.path.abspath('./music/elaina/audio.mp3')

waveform, sample_rate = torchaudio.load(path, format="mp3")

# mfcc
mfcc_func = torchaudio.transforms.MFCC(sample_rate=sample_rate, n_mfcc=40)
mfcc = mfcc_func(waveform)
print(mfcc.shape)
print(mfcc[0][0])

num_channels, num_frames = waveform.shape

fig, ax = plt.subplots(num_channels*2, 1)
for i in range(num_channels):
    ax[i].specgram(waveform[i].numpy(), Fs=sample_rate)
    ax[i+num_channels].plot(waveform[i].numpy())
plt.show()



```
